```python
import seaborn as sns
data = sns.load_dataset('titanic')
data
```

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
```

```python
df = pd.DataFrame(data)
df
```

```python
df1 = df.drop(['embarked','who','adult_male','alive','alone'],axis=1)
df1
```

```python
print(df1['deck'].value_counts())
```

```python
df1.dropna(subset=["deck"],inplace=True)
df1
```

```python
label_enc = LabelEncoder()
df1['embark_town_encoded'] = label_enc.fit_transform(df1['embark_town'])
df1
```

```python
label_enc = LabelEncoder()
df1['sex_encoded'] = label_enc.fit_transform(df1['sex'])
df1
```

```python
ordinal_enc = OrdinalEncoder()
df1['class_encoded'] = ordinal_enc.fit_transform(df1[['class']])
df1
```

```python
embark_town_onehot = pd.get_dummies(df1,columns=['embark_town'],prefix='emb_town',dtype=int)
embark_town_onehot
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
std_scaler = StandardScaler()
df1['fare_std_scaled'] = std_scaler.fit_transform(df1[['fare']])
df1
```

```python
df1['fare_std_scaled'].mean()
```

```python
df1['fare_std_scaled'].median()
```

```python
df1['fare_std_scaled'].std()
```

```python
min_scaler = MinMaxScaler(feature_range=(0,1))
df1['fare_min_scaled'] = min_scaler.fit_transform(df1[['fare']])
df1
```

```python
numerical_columns = ['survived','pclass','sibsp','parch']
scaled_df = std_scaler.fit_transform(df1[numerical_columns])
df1[numerical_columns] = scaled_df
df1[numerical_columns]
```
